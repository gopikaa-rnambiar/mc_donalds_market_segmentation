```python
from sklearn.decomposition import PCA
```

```python
pca=PCA()
md_pca=pca.fit_transform(md_x)
```

```python
md_pca
```

```python
print("Explained variance by each principal component:")
print(pca.explained_variance_ratio_)
```

```python
std_dev = np.sqrt(pca.explained_variance_)
prop_var = pca.explained_variance_ratio_
cum_var = np.cumsum(pca.explained_variance_ratio_)

summary_df = pd.DataFrame({
    'Standard Deviation': std_dev,
    'Proportion of Variance': prop_var,
    'Cumulative Proportion': cum_var
})

print(summary_df)
```

```python
pca_matrix=pd.DataFrame(pca.components_,columns=md_x.columns).T
for i in pca_matrix.columns:
    pca_matrix.rename(columns={i:f"PC{i+1}"},inplace=True)
pca_matrix=pca_matrix*-1
print(pca_matrix)
```

```python
# extracting segments
```

```python
from sklearn.cluster import KMeans
wcss=[]
np.random.seed(1234)
for i in range(2,9):
    k_mean=KMeans(n_clusters=i,n_init=10)
    k_mean.fit(md_x)
    pred_k=k_mean.fit_predict(md_x)
    wcss.append(k_mean.inertia_)
```

```python
pred_k
```

```python
wcss
```

```python
plt.bar(range(2,9),wcss)
plt.ylabel("sum of within cluster distances ")
plt.xlabel("number of segments")
plt.show()
```

```python
df['cluster_label']=k_mean.labels_
df
```

```python
df['Like']=np.where(df['Like']=='I love it!+5','+5',df['Like'])
df['Like']=np.where(df['Like']=='I hate it!-5','-5',df['Like'])
```

```python
df3=pd.concat([md_x,df['Like']],axis=1)
df3
```

```python
like_cluster=pd.crosstab(index=df['cluster_label'],columns=df['Like'])
like_cluster
```

```python
np.random.seed(1234)

import statsmodels.api as sm
x=df3.drop(columns=['Like'])
x=sm.add_constant(x)

y=df3[['Like']].astype(int)
result = sm.OLS(y, x).fit()

print(result.summary())
```

```python
import seaborn as sns
ind=result.params.index
val=result.params.values
sns.barplot(y=ind,x=val)
```
